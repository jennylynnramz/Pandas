```python
## Dependencies and Setup
import pandas as pd


## File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

## Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

school_data_renamed = school_data.rename(columns={"type": "school_type"})

## Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data_renamed, how="left", on=["school_name", "school_name"])

#rename type because its causing me problems furthur down


school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
##number of schools
school_count = len(school_data_complete["school_name"].unique())
school_count
```

```python
#total number of students
student_count = school_data_complete["student_name"].count()
student_count
```

```python
##total budget
total_budget = round(school_data["budget"].sum(), 2)
total_budget
```

```python
##average math score
avg_math = round(school_data_complete["math_score"].mean(), 2)
avg_math
```

```python
##average reading score
avg_reading = round(school_data_complete["reading_score"].mean(), 2)
avg_reading
```

```python
#overall passing rate
passing_rate = round(((avg_math + avg_reading)/2), 2)
passing_rate
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)

math_pass = len(school_data_complete[school_data_complete["math_score"] >= 70])
math_pass

pct_math_pass = round((math_pass / student_count) * 100, 2)

pct_math_pass
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
reading_pass = len(school_data_complete[school_data_complete["reading_score"] >= 70])
reading_pass

pct_reading_pass = round((reading_pass / student_count) * 100, 2)
pct_reading_pass


```

```python
#create data frame to display data
district_data = [{"Schools":school_count, "students":student_count, "Total Budget":"$" + str(total_budget), 
                  "Math Avg":str(avg_math)+"%", "Reading Avg":str(avg_reading)+"%", "Passing Score":str(passing_rate)+"%",
                 "Math Pass %":str(pct_math_pass)+"%", "Reading Pass %":str(pct_reading_pass)+"%"} ]
district_summary= pd.DataFrame(district_data)


district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#School Name
#already exists in school_data
school_summary_data = school_data_renamed
```

```python
#School Type
#already exists in school_data

school_summary_data
```

```python
#Total Students



total_students_per_school = student_data.groupby('school_name')['Student ID'].count()
total_students_per_school
```

```python
#Total School Budget
#already exists in school_data


school_summary_data2 = school_summary_data.sort_values(by=['school_name'])
student_data2 = student_data.sort_values(by=['school_name'])
```

```python
#Per Student Budget
#now I need to do some maths.
school_summary_data2['Per Student Budget'] = round(school_summary_data2.budget / school_summary_data2.size, 2)
school_summary_data2

```

```python
#Average Math Score
#also requires math. Has to be pulled from the other df (student_data)

avg_math_bruh = round(student_data.groupby('school_name')['math_score'].mean(), 2)
avg_math_bruh

math_summ_schools = pd.DataFrame({'school_name':avg_math_bruh.index, 'Average Math Scores':avg_math_bruh.values})    
math_summ_schools

summary = pd.merge(school_summary_data2, math_summ_schools, how="left", on=["school_name", "school_name"])
summary
```

```python
#Average Reading Score
avg_reading_school = round(student_data.groupby('school_name')['reading_score'].mean(), 2)
avg_reading_school

reading_summ_schools = pd.DataFrame({'school_name':avg_reading_school.index, 'Average Reading Scores':avg_reading_school.values})    
reading_summ_schools

summary2 = pd.merge(summary, reading_summ_schools, how="left", on=["school_name", "school_name"])
summary2
```

```python
#% Passing Math  


math_school_pass = student_data[student_data['math_score'] >= 70]
math_school_pass

school2 = round((math_school_pass.groupby('school_name')['math_score'].count()) / total_students_per_school * 100, 2)
school2

math_school_pass_df = pd.DataFrame({'school_name':school2.index, 'students_passing_math':school2.values})    
math_school_pass_df


summary3 = pd.merge(summary2, math_school_pass_df, how="left", on=["school_name", "school_name"])
summary3

```

```python
#% Passing Reading
reading_school_pass = student_data[student_data['reading_score'] >= 70]
reading_school_pass

reading_pct_school = round((reading_school_pass.groupby('school_name')['reading_score'].count()) / total_students_per_school * 100, 2)
reading_pct_school

reading_school_pass_df = pd.DataFrame({'school_name':reading_pct_school.index, 'students_passing_reading':reading_pct_school.values})    
reading_school_pass_df

summary4 = pd.merge(summary3, reading_school_pass_df, how="left", on=["school_name", "school_name"])
summary4

```

```python
#Overall Passing Rate (Average of the above two)

summary4['Overall Passing Rate'] = round(((summary4.students_passing_math + summary4.students_passing_reading) /2), 2)
summary4
```

```python
#Create a dataframe to hold the above results

schools_summary = summary4.rename(columns={"school_name": "School Name", "school_type":"School Type",
                                          "size":"Size", "budget":"Budget", "Per Student Budget":"Budget Per Student",
                                          "students_passing_math":"Percent of Students Passing Math",
                                          "students_passing_reading":"Percent of Students Passing Reading"})
schools_summary
```
