```python
## Dependencies and Setup
import pandas as pd


## File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

## Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

school_data_renamed = school_data.rename(columns={"type": "school_type"})

## Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data_renamed, how="left", on=["school_name", "school_name"])

#rename type because its causing me problems furthur down


school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

```python
##number of schools
school_count = len(school_data_complete["school_name"].unique())
school_count
```

```python
#total number of students
student_count = school_data_complete["student_name"].count()
student_count
```

```python
##total budget
total_budget = round(school_data["budget"].sum(), 2)
total_budget
```

```python
##average math score
avg_math = round(school_data_complete["math_score"].mean(), 2)
avg_math
```

```python
##average reading score
avg_reading = round(school_data_complete["reading_score"].mean(), 2)
avg_reading
```

```python
#overall passing rate
passing_rate = round(((avg_math + avg_reading)/2), 2)
passing_rate
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)

math_pass = len(school_data_complete[school_data_complete["math_score"] >= 70])
math_pass

pct_math_pass = round((math_pass / student_count) * 100, 2)

pct_math_pass
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
reading_pass = len(school_data_complete[school_data_complete["reading_score"] >= 70])
reading_pass

pct_reading_pass = round((reading_pass / student_count) * 100, 2)
pct_reading_pass


```

```python
#create data frame to display data
district_data = [{"Schools":school_count, "students":student_count, "Total Budget":"$" + str(total_budget), 
                  "Math Avg":str(avg_math)+"%", "Reading Avg":str(avg_reading)+"%", "Passing Score":str(passing_rate)+"%",
                 "Math Pass %":str(pct_math_pass)+"%", "Reading Pass %":str(pct_reading_pass)+"%"} ]
district_summary= pd.DataFrame(district_data)
district_summary
```
